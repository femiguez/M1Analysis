```python
import pandas as pd
import pprint
import matplotlib.pyplot as plt
import matplotlib.style as pltstyle
from matplotlib.dates import DateFormatter
from sklearn.metrics import mean_squared_error
```

```python
accola_soy_2018 = pd.read_csv('data/apsim_output/accola_apsim_allout_soy_2018.csv')
accola_soy_noyld = pd.read_csv('data/apsim_output/accola_apsim_allout_noyld_soy_2018.csv')
accola_maize_2019 = pd.read_csv('data/apsim_output/accola_apsim_allout_maize_2019.csv')
accola_maize_noyld = pd.read_csv('data/apsim_output/accola_apsim_allout_noyld_maize_2019.csv')
```

```python
accola_soy_2018.head()
```

```python
fig, axes = plt.subplots(1,1)
pltstyle.use('seaborn-notebook')
axes.plot(accola_soy_2018['leach_no3'], label='Soy Yld Leach', color='blue')
axes.plot(accola_soy_noyld['leach_no3'], label='Soy NoYld Leach', color='orange', linestyle='dashed')
axes.legend()
axes.set_ylabel('NO3 leaching', fontsize=14)
#date_form = DateFormatter("%m/%d")
#axes.xaxis.set_major_formatter(date_form)
axes.set_title('APSIM soybean leaching with and without biomass', fontsize=14)
plt.show()
```

```python
fig, axes = plt.subplots(1,1)
pltstyle.use('seaborn-notebook')
axes.plot(accola_maize_2019['leach_no3'], label='Corn Yld Leach', color='blue')
axes.plot(accola_maize_noyld['leach_no3'], label='Corn NoYld Leach', color='orange', linestyle='dashed')
axes.legend()
axes.set_ylabel('NO3 leaching', fontsize=14)
#date_form = DateFormatter("%m/%d")
#axes.xaxis.set_major_formatter(date_form)
axes.set_title('APSIM maize leaching with and without biomass', fontsize=14)
plt.show()
```
